```python
from numpy import *
```

```python
X = array([0, 1, 2, 3, 4, 5])
y = array([0, 0.8, 0.9, 1, -0.8, -1])
print(X)
print(y)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(X, y, 'o')
```

```python
from scipy.interpolate import *

p1 = polyfit(X, y, 1) #linear fit
print("For linear:", p1)
p2 = polyfit(X, y, 2) #quadratic 
print("For quadratic:", p2)
p3 = polyfit(X, y, 3) #cubic
print("For cubic:", p3)

print(polyval(p1, X), polyval(p2, X), polyval(p3, X))

import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(X, y, 'o')
plt.plot(X, polyval(p1, X), 'r')
plt.plot(X, polyval(p2, X), 'b--')
plt.plot(X, polyval(p3, X), 'm1--')
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(X, y, 'o')
xp = linspace(-2,6,100) #Return evenly spaced numbers over a specified interval
plt.plot(xp, polyval(p1, xp), 'r')
plt.plot(xp, polyval(p2, xp), 'b--')
plt.plot(xp, polyval(p3, xp), 'm1--')
plt.show()
```

```python
yfit = p1[0]*X + p1[1]
print(yfit) #predicted values
print(y) #actual values
```

```python
yresid = y - yfit
SSresid = sum(power(yresid,2))
SStotal = len(y) * var(y)
rsq = 1 - (SSresid / SStotal)
print(rsq)
```

```python
from scipy.stats import *
slope, intercept, r_value, p_value, std_err = linregress(X,y) 
print(slope, intercept, r_value, p_value, std_err)
print(power(r_value,2))
```
