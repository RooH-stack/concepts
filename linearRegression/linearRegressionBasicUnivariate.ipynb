```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('../data/house_prices.csv')
data.head()
```

```python
plt.scatter(data['LotArea'], data['SalePrice'])
plt.title('NYC House pricing')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```

```python
X = data.LotArea[:,np.newaxis]  # Reshape
y = data.SalePrice
X, y
```

```python
# Fitting Simple Linear Regression to the Training set
regressor = LinearRegression()
regressor.fit(X, y)
```

```python
# Predicting the Test set results
y_pred = regressor.predict(X)
y_pred
```

```python
# plotting the best fitting line
# Train Part
plt.scatter(X, y)
plt.plot(X, y_pred, "r-")
plt.title('Housing Price ')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```

```python
'''
Visually prediction e.g. :
If we want to buy a house of 14,000 sq. ft, 
we can simply draw a vertical line from 14,000 
up to our Approximated Trend line and continue 
that line towards the y-axis
'''
# Train Part
plt.scatter(X, y)
plt.plot(X, y_pred, "r-")
plt.title('Housing Price ')
plt.xlabel('Area')
plt.ylabel('Price')
plt.axvline(x=14000,c='g');

'''
a house whose area ~ 14,000 we need to pay ~ 2,00,000-2,25,000
'''
```
